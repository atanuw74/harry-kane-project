```python
import json
import io
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from copy import deepcopy
from urllib.request import urlopen
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Scrape PL stats:
pl_scrape_url = "https://understat.com/league/EPL/2020"
page_connect = urlopen(pl_scrape_url)
page_html = BeautifulSoup(page_connect, "html.parser")

json_raw_string = page_html.findAll(name="script")[3].text
start_ind = json_raw_string.index("\\")
stop_ind = json_raw_string.index("')")

json_data = json_raw_string[start_ind:stop_ind]
json_data = json_data.encode("utf8").decode("unicode_escape")

pl_top_df = pd.json_normalize(json.loads(json_data))
```

```python
pl_top_df
```

```python
# Scrape Bundesliga stats:
bundesliga_scrape_url = "https://understat.com/league/Bundesliga/2020"
page_connect = urlopen(bundesliga_scrape_url)
page_html = BeautifulSoup(page_connect, "html.parser")

json_raw_string = page_html.findAll(name="script")[3].text
start_ind = json_raw_string.index("\\")
stop_ind = json_raw_string.index("')")

json_data = json_raw_string[start_ind:stop_ind]
json_data = json_data.encode("utf8").decode("unicode_escape")

bundesliga_top_df = pd.json_normalize(json.loads(json_data))
```

```python
bundesliga_top_df
```

***JOINT PL BUNDESLIGA STATS***

```python
joint_df = pl_top_df.append(bundesliga_top_df)
joint_df
```

```python
joint_df.info()
```

```python
joint_df = joint_df.apply(pd.to_numeric, errors="ignore")
```

```python
joint_df.info()
```

***REQUIRED DATA***

```python
top = deepcopy(joint_df[joint_df["goals"] > 9])
```

```python
top
```

```python
top.reset_index(drop=True, inplace=True)
```

```python
top
```

```python
top.info()
```

```python
top["Goals_Per_90"] = top["goals"].divide(top["time"]).multiply(90)
```

```python
top["Assists_Per_90"] = top["assists"].divide(top["time"]).multiply(90)
```

```python
top["Shots_Per_90"] = top["shots"].divide(top["time"]).multiply(90)
```

```python
top["Key_Passes_Per_90"] = top["key_passes"].divide(top["time"]).multiply(90)
```

```python
top["xG_Per_90"] = top["xG"].divide(top["time"]).multiply(90)
```

```python
top["xA_Per_90"] = top["xA"].divide(top["time"]).multiply(90)
```

```python
top["npg_Per_90"] = top["npg"].divide(top["time"]).multiply(90)
```

```python
top["npxG_Per_90"] = top["npxG"].divide(top["time"]).multiply(90)
```

```python
top["xGChain_Per_90"] = top["xGChain"].divide(top["time"]).multiply(90)
```

```python
top["xGBuildup_Per_90"] = top["xGBuildup"].divide(top["time"]).multiply(90)
```

```python
per90_stats = ["Goals_Per_90", "Assists_Per_90", "Shots_Per_90", "Key_Passes_Per_90", "xG_Per_90", "xA_Per_90", "npg_Per_90", "npxG_Per_90", "xGChain_Per_90", "xGBuildup_Per_90"]
```

```python
top.loc[(top["id"] == 647), per90_stats].sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
top[per90_stats] = scaler.fit_transform(top[per90_stats])
```

***PLOTTING***

```python
top_max = top[per90_stats].max().max()
top_max
```

```python
import plotly.graph_objects as go
```

```python
for i, row in top.iterrows():
    if row["id"] == 647:
        continue
    # Initiate the plotly go figure resembles a canvas
    fig = go.Figure()
    # Add Radar plots for different players:
    fig.add_trace(go.Scatterpolar(
        r=top.loc[(top["id"] == 647), per90_stats].sum(),
        theta=per90_stats,
        fill='toself',
        name='Harry Kane'))
    fig.add_trace(go.Scatterpolar(
        r=top.loc[(top["id"] == row["id"]), per90_stats].sum(),
        theta=per90_stats,
        fill='toself',
        name=row["player_name"]))
    # Additional properties for the plot:
    fig.update_layout(
        title="Harry Kane vs " + row["player_name"],
    polar=dict(
        radialaxis=dict(
        visible=True,
        range=[0, top_max]
        )),
    showlegend=True
    )
    fig.show()
```
